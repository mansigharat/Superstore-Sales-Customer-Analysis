<h2>Which products contribute most to revenue?</h2>

```python
query = """
SELECT 
Category,
ROUND(SUM(Sales),2)as Total_Sales,
ROUND(SUM(Profit) , 2) as Total_Profits
from Orders
GROUP BY category
"""

df = pd.read_sql(query, conn)
df
```

**INSIGHTS :**
    
- Technology has the highest sales and profit.
- Furniture has moderate sales but very low profit.
- Office Supplies has slightly lower sales but decent profit.

```python
plt.figure(figsize=(8,5))
plt.bar(df['Category'], df['Total_Sales'])
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.title('Total Sales by Product Category')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df['Category'], df['Total_Profits'], color='green')
plt.xlabel('Category')
plt.ylabel('Total Profits')
plt.title('Total Profits by Product Category')
plt.show()
```

<h2> How do sales vary over time? Are there seasonal trends or peak months? </h2>

```python
query = """
SELECT
    YEAR(Order_Date) AS Year,
    MONTH(Order_Date) AS Month_Num,
    DATENAME(MONTH, Order_Date) AS Month_Name,
    ROUND(SUM(Sales),2) AS Total_Sales
FROM Orders
GROUP BY YEAR(Order_Date), MONTH(Order_Date), DATENAME(MONTH, Order_Date)
ORDER BY Year, Month_Num
"""

df_trends = pd.read_sql(query, conn)
df_trends.head(12)
```

```python
monthly_sales = df.groupby(['Year','Month'])['Sales'].sum().reset_index()
monthly_sales.head()
```

**Insights**
- Sales peak in December, likely due to holiday demand.
- February consistently has low sales, indicating off-season.
- Seasonal fluctuations suggest opportunities for marketing campaigns in low months and inventory prep for peak months.

<h2> Who are our most important customers? How often do they buy, and how much do they spend? </h2>

```python
query = """
SELECT 
    Customer_ID,
    Customer_Name,
    COUNT(Order_ID) AS Total_Orders,
    ROUND(SUM(Sales),2) AS Total_Revenue,
    ROUND(SUM(Sales)/COUNT(Order_ID),2) AS Avg_Order_Value
FROM Orders
GROUP BY Customer_ID, Customer_Name
ORDER BY Total_Revenue DESC
"""
df = pd.read_sql(query, conn)
df.head(10)
```

# Insights

- Mitch Willingham is a big-ticket buyer with high AOV but few orders.
- Sean Miller is both a frequent and high-revenue customer.
- Many customers are occasional buyers with low revenue.
- High-frequency, high-revenue customers should be targeted for loyalty programs.

```python
query = """SELECT 
    Customer_ID,
    Customer_Name,
    ROUND(SUM(Sales)/COUNT(Order_ID), 2) AS Avg_Order_Value,
    COUNT(Order_ID) AS Total_Orders,
    ROUND(SUM(Sales), 2) AS Total_Revenue
FROM Orders
GROUP BY Customer_ID,Customer_Name
ORDER BY Avg_Order_Value DESC;"""

df = pd.read_sql(query,conn)
df
```

```python
top_customers = df.sort_values(by='Total_Orders', ascending=False).head(10)
plt.figure(figsize=(21,7))
plt.bar(top_customers['Customer_Name'], top_customers['Total_Orders'], color='orange')
plt.xlabel('Customer Name')
plt.ylabel('Number of Orders')
plt.title('Top 10 Frequent Customers')
plt.show()
```

**Insight:**  
- Mitch Willingham is a big-ticket buyer with a high AOV of 1,751.29 but only 3 purchases.  
- Sean Miller is both a frequent and high-revenue customer, contributing 25,043.05 across 15 orders.  
- Many customers, like Thais Sissman, are occasional buyers with very low total revenue.  
- High-frequency, high-revenue customers should be prioritized for loyalty programs and personalized marketing, while occasional, low-value customers may need engagement campaigns.


<h2> Which products are customers buying and patterns of co-purchase? </h2>


```python
query = """
SELECT 
    Customer_ID,
    Customer_Name,
    Category,
    ROUND(SUM(Sales),2) AS Sales_Per_Customer,
    COUNT(Order_ID) AS Total_Orders_Per_Customer
FROM Orders
GROUP BY Customer_ID, Customer_Name, Category
ORDER BY Sales_Per_Customer DESC
"""

df_customer_products = pd.read_sql(query, conn)
df_customer_products.head(10)
```

#  Insights 

- High-value customers like Sean Miller and Tamara Chand primarily purchase Technology products.
- Customers like Sanjit Chand purchase mostly Office Supplies and make many smaller orders.
- Technology buyers often also purchase Office Supplies, suggesting opportunities for bundling or cross-selling.
- Furniture purchases are mostly occasional and low-value, indicating niche buyers.



**Summary:**
1. Technology is the most profitable and high-selling category.
2. Sales are seasonal, peaking in December and lowest in February.
3. High-frequency, high-revenue customers (e.g., Sean Miller) are key targets for loyalty programs.
4. Big-ticket buyers (e.g., Mitch Willingham) may benefit from personalized offers.
5. Product-customer patterns indicate co-purchase opportunities (Technology + Office Supplies).

**Recommendations:**
- Target high-value, frequent customers with loyalty campaigns.
- Bundle Technology and Office Supplies for cross-selling.
- Prepare inventory for peak months (Nov-Dec) and create promotions for low-sales months (Feb).
- Engage occasional buyers with marketing campaigns to increase purchase frequency.
